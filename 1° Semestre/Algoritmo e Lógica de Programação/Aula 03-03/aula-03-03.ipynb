```python
a = 6
b = 4

# Divisão inteira
c = a // b
# Resultado = 1.

# Resto da Divisão
d = a % b
# Resultado = 2.

print(f"A divisão inteira de {a} por {b} é: {c}")
print(f"A resto de {a} por {b} é: {d}")
```

```python
e = 12
f = 2

# Soma
print(f"Soma: {e + f}")

# Subtração
print(f"Subtração: {e - f}")

# Multiplicação
print(f"Multiplicação: {e * f}")

# Divisão
print(f"Divisão: {e / f}")

# Exponenciação
print(f"Exponenciação: {e ** f}") 
```
