```python
# Exercício 1 - Crie um algoritmo em Python para calcular a média aritmética das notas de 40 alunos.

notas = []

for i in range(1, 41):
    # Isso aqui serve só para quando clicar em "Executar Tudo", ele não tenha que executar tudo, e pula essa parte. Se estiver comentado, executa o código normalmente.
    # break
    for quant_notas in range(1, 4):
        while True:
            try:
                nota = float(input(f"Digite a {quant_notas}ª nota do {i}° aluno: "))
                if nota > 10 or nota < 0:
                    print("Erro: Digite uma nota válida de 0 a 10.")
                else:
                    notas.append(nota)
                    break
                    # Aqui ele quebra o andamento padrão e sai do While
            except ValueError:
                print(f"Erro: Impossível de ler valores desse tipo. Digite uma nota válida de 0 a 10. (Use pontos(.) para valores decimais")
                # Aqui ele quebra o andamento do try e volta para o While
    media = sum(notas) / len(notas)
    print(f"O(a) aluno(a) de número {i} teve média aritmética de: {media:.2f}.")
    notas.clear()
```

```python
# Exercício 2 - Desenvolva um algoritmo em Python para calcular a soma dos 100 primeiros números inteiros (1 + 2 + 3... + 98 + 99 + 100).

numeros = []

for num in range (1, 101):
    numeros.append(num)

soma = sum(numeros)
print(f"O resultado da soma dos 100 primeiros números é: {soma}")
```

```python
# Exercício 3 - Crie um algorimo em python para calcular a média aritmética de 4 notas de um aluno. Se a média for maior ou igual a 6.0, o aluno está APROVADO direto. Se a média ficar entre 3.0 e 5.9, o aluno será informado que poderá fazer uma prova de EXAME, caso contrário, estará REPROVADO.

notas = []

for i in range(1, 5):

    # Fica repetindo até houver uma quebra voluntária.
    while True:
        try:
            nota = float(input(f"Digite a {i}ª nota: ").replace(',', '.'))
            # Verifica se o valor digitado está entre 0 e 10. 
            if 0 <= nota <= 10:
                # Adiciona nota para o array notas.
                notas.append(nota)
                break
            else:
                print("Erro: Digite um número válido de 0 a 10.")
        # Verifica se o valor foi digitado corretamente 
        except ValueError:
            print("Erro: Impossível de ler os valores. Digite um número válido de 0 a 10.")

media = sum(notas) / len(notas)
if media >= 6:
    print(f"Você foi APROVADO direto. Sua média foi de: {media}")
elif 3 < media < 6:
    print(f"Você poderá fazer um EXAME de admissão. Sua média foi de: {media}")
else:
    print(f"Você foi REPROVADO. Sua média foi de: {media}")
```
