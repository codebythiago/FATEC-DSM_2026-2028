```python
# Estruturas de Repetição

# Estrutura FOR IN RANGE

for num in range (10):
    print(num)
```

```python
for numero in range (1, 10):
    print(numero)
```

```python
for numero in range (1, 11):
    print(numero)
```

```python
for i in range (1, 10, 3):
    print(i)
```

```python
for a in range (-20, -3, 2):
    print(a)
```

```python
for a in range (-4, -20, -2):
    print(a)
```
