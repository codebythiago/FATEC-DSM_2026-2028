```python
# Comandos de Entrada
nome = input("Digite o seu nome: ")
curso = input("Digite o nome do seu curso: ")
# nota1 = float(input("Digite sua primeira nota: "))
# nota2 = float(input("Digite sua segunda nota: "))
# nota3 = float(input("Digite sua terceira nota: "))
# nota4 = float(input("Digite sua quarta nota: "))
# OOOOUUU
nota1 = eval(input("Digite sua primeira nota: "))
nota2 = eval(input("Digite sua segunda nota: "))
nota3 = eval(input("Digite sua terceira nota: "))
nota4 = eval(input("Digite sua quarta nota: "))

# EVAL: ele avalia o valor do dado digitado... Se digitou inteiro, recebe inteiro. Se digitou string, recebe str. Se digitou float, recebe float.
# EX.: Se digitou 5, -5, 42, 0, recebe int.
# EX.: Se digitou Thiago, Evanivaldo, Python, Estudo, recebe str.
# EX.: Se digitou 2.52, -4.3, 3.0 recebe float.
# EX.: Se digitou true, TRUE, false, FALSE, recebe bool.

# Calculando a média
media = (nota1 + nota2 + nota3 + nota4)/4
```

```python
# Comando de Decisão
if media >= 6:
    # Comando de Saída
    print(f"O aluno {nome} do curso de {curso} foi APROVADO com média: {media}.")
elif media < 3:
    # Comando de Saída
    print(f"O aluno {nome} do curso de {curso} foi REPROVADO com média: {media}.")
else:
    # Comando de Saída
    print(f"O aluno {nome} do curso de {curso} tirou média {media} e pode fazer o EXAME DE ADMISSÃO.")
```
